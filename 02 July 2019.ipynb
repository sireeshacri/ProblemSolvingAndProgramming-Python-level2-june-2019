```python
# Function to read csv data then print it
import pandas as pd
def readCSVfile(filepath):
    return pd.read_csv(filepath)

filepath='DataFiles/Income.csv'
readCSVfile(filepath)
```

```python
incomedf = readCSVfile(filepath)

# Function to print all columns
def printDataFrameColumms(df):
    columns=df.columns
    for column in columns:
        print(column,end=' ')
    return

printDataFrameColumms(incomedf)
```

```python
# Function to access entire row based on a unique value in row
def printUniqueRow(df,key):
    for row in df.values:
        if key in row:
            for item in row:
                print(item,end=' ')
            print('\n')
    return

printUniqueRow(incomedf,'Alaska')
```

```python
# Accessing a unique value based on row, column info
# Income of a state in a given year

def getRowIndex(df,rowkey):
    for i in range(len(df.values)):
        if df.values[i][0]==rowkey or df.values[i][1]==rowkey:
            rowindex = i
    return rowindex
    
getRowIndex(incomedf,'Arizona')
```

```python
def getColumnIndex(df,columnkey):
    for i in range(len(df.columns)):
        if df.columns[i]==columnkey:
            columnindex=i
    return columnindex

#getColumnIndex(incomedf,'2006')
```

```python
def valueRowColumn(df,rowkey,columnkey):
    rowindex=getRowIndex(df,rowkey)
    columnindex=getColumnIndex(df,columnkey)
    return df.values[rowindex][columnindex]
valueRowColumn(incomedf,'Alaska','2006')
    
```

```python
# Function to update data based on rowkey aRnd columnkey
def updateDataFromRowColumn(df,rowkey,columnkey,newdata):
    rowindex=getRowIndex(df,rowkey)
    print(rowindex)
    columnindex=getColumnIndex(df,columnkey)
    row=df.values[rowindex]
    row[columnindex]= newdata
    df.loc[rowindex]=row
    return 
    
updateDataFromRowColumn(incomedf,'Arizona','2006',46657)
incomedf
```

```python
# Function to get starting year index
def startIndex(df,startyear):
    for i in range(len(df.columns)):
        if df.columns[i] == startyear:
            startindex = i
    return startindex

#startIndex(incomedf,'2007')
    
```

```python
def endIndex(df,columnkey):
    for i in range(len(df.columns)):
        if df.columns[i]==columnkey:
            endindex=i
    return endindex
#endIndex(incomedf,'2010')
```

```python
# Function to State with lowest average income from 2007 to 2010(inclusive)
def StateLowestAvgInRange(df,startyear,endyear):
    startindex=startIndex(df,startyear)
    endindex=endIndex(df,endyear)
    li={}
    lis=[]
    for row in df.values:
        sum = 0
        count = 0
        for i in range(startindex,endindex+1):
            sum +=row[i]
            count +=1
        avg = sum/count
        #print(avg)
        li[row[1]]=avg
        lis.append(avg)
    print(li)
    print(lis)
    for items in li.items():
        if min(lis)==items[1]:
            print("lowest avg state is",items[0],':',items[1])
    return

                
StateLowestAvgInRange(incomedf,'2007','2010')
```
